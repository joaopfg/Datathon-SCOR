```r
library(survival)
```

```r
set.seed(0)
```

```r
df <- read.csv("Final_NHANES.csv")
```

```r
colnames(df)[1] <- "id"
```

```r
for (i in 1:nrow(df)){
    df$End[i] = df$Age[i] + 2
    df$id[i] = df$id[i] - 51623
}
```

```r
levels(df$mortstat) <- c(0,1)
```

```r
head(df)
```

```r
new_df <- df[,c("id","Age", "End", "BPSysAve", "DirectChol", "BPDiaAve", "Weight", "PhysActive" , "Poverty", "TotChol", "UrineFlow1", "DaysMentHlthBad", "UrineVol1", "DaysPhysHlthBad", "BMI", "HHIncomeMid", "SleepHrsNight", "Pulse", "Gender", "Diabetes", "Smoke100", "SleepTrouble", "mortstat")]
```

```r
LTRCIT.fit <- LTRCIT(Surv(Age, End, mortstat) ~ BPSysAve + DirectChol + BPDiaAve + Weight + PhysActive + Poverty + TotChol + UrineFlow1 + DaysMentHlthBad + UrineVol1 + DaysPhysHlthBad + BMI + HHIncomeMid + SleepHrsNight + Pulse + Gender + Diabetes + Smoke100 + SleepTrouble, new_df)
LTRCART.fit <- LTRCART(Surv(Age, End, mortstat) ~ BPSysAve + DirectChol + BPDiaAve + Weight + PhysActive + Poverty + TotChol + UrineFlow1 + DaysMentHlthBad + UrineVol1 + DaysPhysHlthBad + BMI + HHIncomeMid + SleepHrsNight + Pulse + Gender + Diabetes + Smoke100 + SleepTrouble, new_df)
```
